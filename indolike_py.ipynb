```python

//CALCULATOR//
def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def multiply(a,b):
    return a*b

def divide(a,b):
    if b== 0:
        return "Error:Division by zero is Undefined"
    return a/b

def calculator():
    print("Welcome User (: ")
    print("Available Operations: ")
    print("'+' for  Addition")
    print("'-' for Substraction")
    print("'*' for Multiplication")
    print("'/' for Division")
    while True:
        operation = input("Enter the type of operation ('+','-','*','/') :")
        if operation in ('+','-','*','/'):
            try:
                num1 = float(input("Enter First Number:"))
                num2 = float(input("Enter Second Number:"))
            except valueError:
                print("Invalid input,please enter a valid input...")
                continue
            if operation == '+':
                print(f"The addition of given numbers is: {add(num1,num2)}")
            elif operation == '-':
                print(f"The subraction of given numbers is: {sub(num1,num2)} ")
            elif operation == '*':
                print(f"The multiplication of given numbers is: {multiply(num1,num2)}")
            elif operation == '/':
                print(f"The division of given number is: {divide(num1,num2)}")
            continue_calculation = input("Do you want to perform another calculation?(yes/No):")
            if continue_calculation.lower() != "yes":
                break
        else:
                print("Invalid input,please enter valid inputs......")
if __name__ == "__main__":
    calculator()

```

```python
def display_menu():
    print("\n To-Do Manager")
    print("press: '1' To View task" )
    print("press: '2' To Add task")
    print("press: '3' To Update task ")
    print("press: '4' To Delete task")
    print("press: '5' To Exit")


def view_tasks(tasks):
    if not tasks:
        print("No tasks available.")
    else:
        print("\nTasks:")
        for task_id, task in tasks.items():
            status = "✓" if task["completed"] else "✗"
            print(f"[{status}] ID: {task_id} - {task['description']}")

def add_task(tasks,next_id):
    description = input("Enter the description: ")
    tasks[next_id] = {"description":description,"completed":False}
    print(f"Task added with ID:{next_id}")
    return next_id +1




def delete_task(tasks):
    try:
        task_id = int(input("Enter task ID to delete: "))
        if task_id in tasks:
            del tasks[task_id]
            print("Task deleted successfully.")
        else:
            print("Task ID not found.")
    except ValueError:
        print("Invalid input.")

def update_task(tasks):
    try:
        task_id = int(input("Enter task ID to update: "))
        if task_id in tasks:
            description = input("Enter new description (leave blank to keep current): ")
            completed_input = input("Mark as completed? (yes/no/leave blank): ").lower()
            if description:
                tasks[task_id]["description"] = description
            if completed_input == "yes":
                tasks[task_id]["completed"] = True
            elif completed_input == "no":
                tasks[task_id]["completed"] = False
            print("Task updated successfully.")
        else:
            print("Task ID not found.")
    except ValueError:
        print("Invalid input.")





def main():
    tasks = {}
    next_id = 1
    while True:
        display_menu()
        choice = input("Enter your choice..")
        if choice == "1":
            view_tasks(tasks)
        elif choice == "2":
            next_id = add_task(tasks, next_id)
        elif choice == "3":
            update_task(tasks)
        elif choice == "4":
            delete_task(tasks)
        elif choice == "5":
            print("Goodbye! (: )")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()





```

```python
from PIL import Image, ImageDraw, ImageFont, ImageFilter

def open_image(file_path):
    try:
        img = Image.open(file_path)
        print("Image loaded successfully!")
        return img
    except Exception as e:
        print(f"Error loading image: {e}")
        return None

def resize_image(img, width, height):
    return img.resize((width, height))

def rotate_image(img, angle):
    return img.rotate(angle, expand=True)

def add_text_to_image(img, text, position, font_size=30, color="white"):
    try:
        draw = ImageDraw.Draw(img)
        font = ImageFont.truetype("arial.ttf", font_size)  # Ensure arial.ttf is available
        draw.text(position, text, fill=color, font=font)
        return img
    except Exception as e:
        print(f"Error adding text: {e}")
        return img

def apply_filter(img, filter_type):
    if filter_type == "BLUR":
        return img.filter(ImageFilter.BLUR)
    elif filter_type == "CONTOUR":
        return img.filter(ImageFilter.CONTOUR)
    elif filter_type == "DETAIL":
        return img.filter(ImageFilter.DETAIL)
    elif filter_type == "SHARPEN":
        return img.filter(ImageFilter.SHARPEN)
    else:
        print("Invalid filter type. No filter applied.")
        return img

def save_image(img, save_path):
    try:
        img.save(save_path)
        print(f"Image saved successfully at {save_path}")
    except Exception as e:
        print(f"Error saving image: {e}")

if __name__ == "__main__":
    print("Welcome to the Image Editor!")

    file_path = input("Enter the path to your image: ")
    image = open_image(file_path)

    if image:
        while True:
            print("\nChoose an operation:")
            print("1. Resize Image")
            print("2. Rotate Image")
            print("3. Add Text to Image")
            print("4. Apply Filter (BLUR, CONTOUR, DETAIL, SHARPEN)")
            print("5. Save Image")
            print("6. Exit")

            choice = input("Enter your choice: ")

            if choice == "1":
                width = int(input("Enter new width: "))
                height = int(input("Enter new height: "))
                image = resize_image(image, width, height)
                print("Image resized successfully.")

            elif choice == "2":
                angle = int(input("Enter rotation angle (in degrees): "))
                image = rotate_image(image, angle)
                print("Image rotated successfully.")

            elif choice == "3":
                text = input("Enter the text to add: ")
                x = int(input("Enter the x-coordinate: "))
                y = int(input("Enter the y-coordinate: "))
                font_size = int(input("Enter the font size: "))
                color = input("Enter the text color (default is white): ") or "white"
                image = add_text_to_image(image, text, (x, y), font_size, color)
                print("Text added successfully.")

            elif choice == "4":
                filter_type = input("Enter the filter type (BLUR, CONTOUR, DETAIL, SHARPEN): ").upper()
                image = apply_filter(image, filter_type)
                print(f"Filter '{filter_type}' applied successfully.")

            elif choice == "5":
                save_path = input("Enter the path to save the edited image: ")
                save_image(image, save_path)

            elif choice == "6":
                print("Exiting the Image Editor. Goodbye!")
                break

            else:
                print("Invalid choice. Please try again.")
```
